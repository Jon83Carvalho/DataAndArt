```python
import pandas as pd
```

```python
stock_df=pd.read_csv('day1_1.csv')
```

```python
stock_df
```

```python
stock_df['price_reduced']=stock_df['price'].round(1)
```

```python
stock_df
```

```python
stock_df_g=stock_df.groupby(['Symbol','price_reduced','size']).count()
```

```python
stock_df_g
```

```python
stock_df_g=stock_df_g[['timestamp']].sort_values(by=['timestamp'])
```

```python
stock_df_g=stock_df_g.rename_axis(['Symbol','price','size'])
```

```python
stock_df_g=stock_df_g.rename(columns={'price':'PR'})
```

```python
stock_df_g
```

```python
stock_df_g.to_csv('day1_g_1.csv')
```
