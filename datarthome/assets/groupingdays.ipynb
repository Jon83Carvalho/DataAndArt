```python
import pandas as pd
```

```python
import numpy as np
```

```python
stock_df=pd.read_csv('day1_2.csv')
```

```python
stock_df
```

```python
stock_df['price_reduced']=stock_df['price'].round(0)
```

```python
stock_df
```

```python
max_size=stock_df['size'].max()
```

```python
max_size
```

```python
stock_df_g=stock_df.groupby(by=['Symbol','size',pd.cut(stock_df["price"], np.arange(0, 115, 15))]).agg({'price':'mean','timestamp':'count'})
```

```python
stock_df_g
```

```python
stock_df_g=stock_df_g.dropna()
```

```python
stock_df_g['timestamp'].max()
```

```python
stock_df_g=stock_df_g.rename_axis(index={'price':'p'})
```

```python
stock_df_g
```

```python
stock_df_g.to_csv('day1_g_2.csv')
```
